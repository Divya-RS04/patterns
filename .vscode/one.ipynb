PATTERNS

```python
r=4
for i in  range(0,4):
    for k in range (4,0,-1):
        print("" , end ="")
    for j in range(0,i*2+1):
        print("*",end="")
    print()
r-=1
```

```python
r=1
for i in  range(4,0,-1):
    for k in range (1,r+1):
        print("" , end ="")
    for j in range(i*2-1,0,-1):
        print("*",end="")
    print()
r+=1
```

```python
for i in range(1, 5):
    for j in range(1, i): 
        print(" ", end="")
    for j in range(1, 5):
        print("*", end="")
    print()
```

```python
for i in range(1, 5):
    for j in range(1, 7):
        if i==1 or i==4 or j==1 or j==6:
            print("*", end= "")
        else:
            print(" ",end = "")
    print()         
```

```python
for i in range(1,5):
    for j in range(1, 7):
        if i==1 or i==4 or j==1:
            print("*", end= "")
        else:
            print(" ",end = "")
    print()         
```
